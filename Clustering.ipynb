# Segmentation

Selanjutnya akan dilakukan segmentasi untuk mengidentifikasi kelompok-kelompok individu dengan karakteristik yang serupa. Dalam kasus ini, kami mengelompokkan individu berdasarkan tingkat stres, rasa kesepian, kepercayaan terhadap pemerintah, kekhawatiran terhadap virus corona, dan kepatuhan terhadap aturan pencegahan.

```python
#Mengelompokkan berdasarkan tingkat stres
pss_col = ['Scale_PSS10_UCLA_1', 'Scale_PSS10_UCLA_2', 'Scale_PSS10_UCLA_3', 'Scale_PSS10_UCLA_4', 'Scale_PSS10_UCLA_5', 'Scale_PSS10_UCLA_6', 'Scale_PSS10_UCLA_7', 'Scale_PSS10_UCLA_8', 'Scale_PSS10_UCLA_9', 'Scale_PSS10_UCLA_10']
stress_df['avg_PSS10'] = stress_df[pss_col].mean(axis=1)
stress_df.drop(pss_col, axis=1, inplace=True)

#Mengelompokkan berdasarkan rasa kesepian
lon_col = ['Scale_SLON_1', 'Scale_SLON_2', 'Scale_SLON_3']
stress_df['avg_lon'] = stress_df[lon_col].mean(axis=1)
stress_df.drop(lon_col, axis=1, inplace=True)

#Mengelompokkkan berdasarkan kepercayaan terhadap lingkungan (Orang sekitar dan Pemerintah)
oecd_col = ['OECD_people_1', 'OECD_people_2', 'OECD_insititutions_1', 'OECD_insititutions_2', 'OECD_insititutions_3', 'OECD_insititutions_4', 'OECD_insititutions_5', 'OECD_insititutions_6']
stress_df['avg_oecd'] = stress_df[oecd_col].mean(axis=1)
stress_df.drop(oecd_col, axis=1, inplace=True)

#Mengelompokkan berdasarkan Ketakutan terhadap Virus Corona
concern_col = ['Corona_concerns_1', 'Corona_concerns_2', 'Corona_concerns_3', 'Corona_concerns_4', 'Corona_concerns_5']
stress_df['avg_concern'] = stress_df[concern_col].mean(axis=1)
stress_df.drop(concern_col, axis=1, inplace=True)

#Mengelompokkan berdasarkan Skor kepatuhan kepada peraturan
comp_col = ['Compliance_1', 'Compliance_2', 'Compliance_3', 'Compliance_4', 'Compliance_5', 'Compliance_6']
stress_df['avg_comp'] = stress_df[comp_col].mean(axis=1)
stress_df.drop(comp_col, axis=1, inplace=True)
```

```python
country_map = {'United States': 0, 'United Kingdom': 1, 'Ireland': 2, 'Canada': 3}
stress_df['country'] = stress_df['Country'].map(country_map)
stress_df.drop('Country', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
num_col =  ['avg_PSS10', 'avg_lon', 'avg_oecd', 'avg_concern', 'avg_comp', 'Trust_countrymeasure']
df_std = scaler.fit_transform(stress_df[num_col])
df_std = pd.DataFrame(df_std, columns=num_col, index=stress_df.index)
df_stress= stress_df.copy()
stress_df[num_col] = df_std
```

Akan digunakan Elbow Method untuk menentukan jumlah klaster yang optimal dalam proses clustering (dalam hal ini, k-means).


```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(stress_df)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

Pada grafik tersebut, sumbu X merepresentasikan jumlah klaster (k), sementara sumbu Y menunjukkan Cost Sum (CS), yang merupakan jarak kuadrat dari setiap titik data ke centroid terdekat. Dari grafik tersebut, terlihat bahwa penurunan nilai CS mulai melambat setelah klaster k = 2 yang berarti penurunan setelah titik tersebut tidak signifikan. Artinya, dua klaster mungkin sudah cukup untuk mengelompokkan data secara efisien.


```python
kmeans1 = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
kmeans1.fit(stress_df)
```

```python
df_stress["Segment K-means with k=2"] = kmeans1.labels_
```

```python
df_segm_analysis = df_stress.groupby(['Segment K-means with k=2']).mean()
df_segm_analysis.drop(['country'], axis=1, inplace=True)
df_segm_analysis
```

Kemudian, interpretasi per-klasternya sebagai berikut:

Nilai avg_concern juga relatif tinggi (4.55), menunjukkan kekhawatiran yang lebih besar terhadap virus corona dibanding klaster lain. Mereka cenderung memiliki tingkat stres yang lebih rendah (avg_PSS10 sebesar 2.88) dibandingkan dengan Klaster 1.
